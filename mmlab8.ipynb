```python
import numpy as np
import math
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
p_cr=35
p1=13.3
p2=14.5
tau1=35
tau2=30
N=93
q=1
```

```python
x0=np.array([8,9])
t=np.arange(0,30,0.01)
```

```python
a1=p_cr/(tau1*tau1*p1*p1*N*q)
a2=p_cr/(tau2*tau2*p2*p2*N*q)
b=p_cr/(tau1*tau1*tau2*tau2*p1*p1*p2*p2*N*q)
c1=(p_cr-p1)/(tau1*p1)
c2=(p_cr-p2)/(tau2*p2)
```

```python
def syst1(x,t):
    dx1=x[0]-(b/c1)*x[0]*x[1]-(a1/c1)*x[0]*x[0]
    dx2=(c2/c1)*x[1]-(b/c1)*x[0]*x[1]-(a2/c1)*x[1]*x[1]
    return [dx1, dx2]
```

```python
y1=odeint(syst1, x0, t)
```

```python
plt.plot(t,y1,lw=2)
plt.title('Случай 1')
plt.grid('True')
```

```python
def syst2(x,t):
    dx1=x[0]-((b/c1)+0.00018)*x[0]*x[1]-(a1/c1)*x[0]*x[0]
    dx2=(c2/c1)*x[1]-(b/c1)*x[0]*x[1]-(a2/c1)*x[1]*x[1]
    return [dx1, dx2]
```

```python
y2=odeint(syst2, x0, t)
```

```python
plt.plot(t,y2,lw=2)
plt.title('Случай 2')
plt.grid('True')
```
